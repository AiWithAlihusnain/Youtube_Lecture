```python
marks =  int(input ("Please Enter your marks "))
if marks >= 90:
    print("Grade A")
elif marks >=70:
    print("Grade B")
elif marks >= 50:
    print("Grade c")
else:
    print("Fail")
print(type(marks))
print(marks)
```

```python
signal = input ("enter signals")
if signal == "red":
    print("stop")
elif signal == "green":
    print("Go")
elif signal == "yellow":
    print("Slow down")
else:
    print("Error Please input red green or yellow")
```
