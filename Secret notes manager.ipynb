# Secret Notes Manager 
1. Notes likhy ga
2. Note padhy ga
3. Best practice use krega 

```python
def save_notes():
    note = input ("Apna Secret note likho: ")
    with open("secret_note.txt", "a") as file:
        file.write(note + "\n")
    print("Note saved successfully!")
```

```python
def view_notes():
    try:
        with open("secret_note.txt", "r") as file:
            note = file.read()
            print("Yeh rahe tumhare secret notes:")
            print(note)
    except FileNotFoundError:
        print("Koi notes nahi mile. Pehle kuch notes banao!")
```

```python
while True:
    print("\n1. Notes Likho")
    print("2. Notes Dekho")
    print("3. Exit Karo")
    choice = input("Apna choice chuno (1/2/3): ")
    if choice == '1':
        save_notes()
    elif choice == '2':
        view_notes()
    elif choice == '3':
        print("Bye! Apna khayal rakhna.")
        break
    else:
        print("Invalid choice, dobara try karo.")
        

```
