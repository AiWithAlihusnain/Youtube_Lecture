# Student Result Management System 

# Tuple to Fixed subjects

```python
subjects = ("math" , "Physics" , "Chemistry" , "Computer" , "English" )
```

```python
student_name = input("Enter Student name: ")

```

# use list to store marks

```python
marks = []
for subject in subjects: 
    score = int(input(f"Enter marks for {subject}: "))
    marks.append(score)
    
```

# Calculate percentage

```python
total_marks = 500
obtained_marks = sum(marks)
percentage = obtained_marks / total_marks * 100
```

# Grade System

```python
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:  
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 40:
    grade = "D"
else:
    grade = "F"
```

# Display result

```python
print ("\n-----Student Report Card-----")
print(f"Student Name       :       {student_name}")
print(f"Total Marks        :       {total_marks}")
print(f"Obtained Marks     :       {obtained_marks}")
print(f"Percentage         :       {percentage:.2f}%")
print(f"Grade              :       {grade}")
```
