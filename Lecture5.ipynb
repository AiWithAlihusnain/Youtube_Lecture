# List 

```python
number = [23,45,56,78]
print(number)
```

# apppend()

```python
number.append(90)
```

```python
print(number)
```

# pop()

```python
number.pop()
print(number)
```

# Sort()

```python
prices = [400, 800, 100, 400, 200]
prices.sort()
print(prices)
```

# len()

```python
print(len(prices))
```

# slicing

```python
print(prices[1:7])
```

# List Comprehension

```python
prices = [400, 800, 100, 400, 200]
squares = [i**2 for i in prices]
print(squares)
```

```python
number = [23,45,56,78]
for i in number:
    print(i)
```

# Practical Example 

Average = sum of all values divided by number of values

```python
marks = [85, 99, 92, 78, 96, 88, 91]
average = sum(marks) // len(marks)
print("Average marks:", average)
```

```python
for i in range (1, 10):
    if i % 2 == 0:
        print(f"{i} is even")
    else:
        print(f"{i} is odd")    
```

```python
even_numbers = [i for i in marks if i % 2 == 0]
print("Even marks:", even_numbers)
```
