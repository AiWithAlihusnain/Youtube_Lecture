# File Handling 
# open(mode) open the file 
mode = 'r' read Read 
mode = 'w' write but previous data erase
mode = 'a' append data into end 


```python
f = open ("emails.csv", "r")
data = f.readlines()
print(data)
f.close()
```

```python
f = open ("emails.csv")
data = f.read()
print(data)
f.close()
```

```python
f = open ("notes.txt", "w")
f.write("Welcome to day 8 of Python programming.")
f.write("\nWe are learning file handling.")
f.close()
```

```python
f = open ("notes.txt", "w")
f.write("Welcome to day 8 of Python programming.")
f.write("\nWe are learning file handling.")
f.close()
```

```python
f = open ("notes.txt", "a")
f.write("\nThanku.")
f.close()
```

```python
with open ("notes.txt", "r") as f:
    data = f.read()
    print(data)
```

```python
import csv
data = [["Name", "Age"], ["Alice", 30], ["Bob", 25], ["Charlie", 35]]
with open ("student.csv", "w" ) as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
with open ("student.csv", "r") as f:
    content = f.read()
    print(content)
```

```python
try:
    with open("student.txt", "r") as f:
        data = f.read()
        print(data)
except FileNotFoundError:
    print("The file does not exist.")
```
